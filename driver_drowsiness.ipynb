```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import os
```

```python
data_directory = "mrlEyes/Train_Eyes/"
classis = ['Closed_Eyes','Open_Eyes']
img_size = 224
training_data=[]
def create_data():
    for category in classis:
        path = os.path.join(data_directory,category)
        img_num = classis.index(category)
        for img in os.listdir(path):
            img_array = cv.imread(os.path.join(path,img),cv.IMREAD_GRAYSCALE)
            img_to_color = cv.cvtColor(img_array,cv.COLOR_GRAY2RGB)
            img_resize = cv.resize(img_to_color,(img_size,img_size))
            print(img_num)
            training_data.append([img_resize,img_num])

create_data()
```

```python
len(training_data)
```

```python
import random

random.shuffle(training_data)
```

```python
x=[]
y=[]
def split_data_label():
    for features,label in training_data:

        x.append(features)
        y.append(label)
        print("ss")

    x = np.array(x).reshape(-1,img_size,img_size,3)
    y = np.array(y)
    return x,y
```

```python
x = x/255.0; # normalize x
```

```python
def save_data(x,y):

    pickle_out = open('x.pickle','wb')
    pickle.dump(x,pickle_out)

    pickle_out = open('y.pickle','wb')
    pickle.dump(y,pickle_out)
```

```python
import pickle

def load_data()
    pickle_in = open('x.pickle','rb')
    x = pickle.load(pickle_in)

    pickle_in = open('y.pickle','rb')
    y = pickle.load(pickle_in)
    
    return x,y
```

Load model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
base_outout = model.layers[-4].output

flat_layer = layers.Flatten()(base_outout)
final_output = layers.Dense(1)(flat_layer)
final_ouput = layers.Activation('sigmoid')(final_output)
```

```python
new_model = keras.Model(inputs=base_input,outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
new_model.fit(x=x,y=y,epochs=3,validation_split=0.1)
```

```python
new_model.save('my_model.h5')
```
